```python
import pandas as pd
import numpy as np
import plotly.graph_objects as go
```

```python
# Load the dataset
df = pd.read_csv('English_Abstractive.csv')


def split_classes(df):
    # Split the dataset into classes based on 'Judgement Status'
    df_class1 = df[df['Judgement Status'] == 0]
    df_class2 = df[df['Judgement Status'] == 1]
    df_class3 = df[df['Judgement Status'] == 2]
    df_class4 = df[df['Judgement Status'] == 3]
    return df_class1, df_class2, df_class3, df_class4

df_class1, df_class2, df_class3, df_class4 = split_classes(df)

def calculate_mean_vectors(df_class1, df_class2, df_class3, df_class4):
    # Calculate the mean vector for each class (excluding the last column)
    mean_class1 = np.mean(df_class1.iloc[:, :-1])
    mean_class2 = np.mean(df_class2.iloc[:, :-1])
    mean_class3 = np.mean(df_class3.iloc[:, :-1])
    mean_class4 = np.mean(df_class4.iloc[:, :-1])
    return mean_class1, mean_class2, mean_class3, mean_class4
print(calculate_mean_vectors(df_class1, df_class2, df_class3, df_class4))
```

```python
def calculate_mean_vectors_per_feature(df_class):
    # Create a list to store mean values for each feature in the class
    mean_vector = [] 
    for i in range(len(df_class.columns) - 1):
        mean_value = np.mean(df_class.iloc[:, i])  # Calculate mean for each feature
        mean_vector.append(mean_value) 
    return np.array(mean_vector)  # Convert the list to a numpy array
print(calculate_mean_vectors_per_feature(df_class1))
```

```python
def calculate_standard_deviations(df_class):
    # Create a list to store standard deviation values for each feature in the class
    std_vector = [] 
    for i in range(len(df_class.columns) - 1):
        std_value = np.std(df_class.iloc[:, i])  # Calculate standard deviation for each feature
        std_vector.append(std_value) 
    return np.array(std_vector)  # Convert the list to a numpy array
print(calculate_standard_deviations(df_class1))
```

```python
def plot_histogram(df_class, feature_column):
    # Choose a feature (column) for histogram analysis
    feature_data = df_class[feature_column]
    
    # Define the number of bins for the histogram
    bins = 10
    
    # Calculate the histogram data
    hist, bin_edges = np.histogram(feature_data, bins=bins)
    
    # Calculate mean and variance for the selected feature
    mean_feature = np.mean(feature_data)
    var_feature = np.var(feature_data)
    
    # Plot the histogram for the selected feature
    plt.figure(figsize=(8,8))
    plt.hist(feature_data, bins=bins)
    plt.xlabel('Feature Value')
    plt.ylabel('Frequency')
    plt.title('Histogram')
    plt.show()

plot_histogram(df_class1, '0')
```

```python
def calculate_minkowski_distance(df_class, feature1, feature2):
    # Select two feature vectors from the class for Minkowski distance calculation
    feature1_vector = np.array(df_class[feature1])
    feature2_vector = np.array(df_class[feature2])
    
    # Function to calculate Minkowski distance
    def min_dis(a, b, r):
        return np.sum(np.abs(a - b) ** r) ** (1 / r)
    
    # Calculate Minkowski distances for r values from 1 to 10
    r_values = np.arange(1, 11)
    distances = [min_dis(feature1_vector, feature2_vector, r) for r in r_values]
    
    # Plot Minkowski distances as a function of r
    plt.figure(figsize=(8,8))
    plt.plot(r_values, distances, marker='o')
    plt.xlabel('r')
    plt.ylabel('Minkowski Distance')
    plt.title('Minkowski Distance between 2 Feature Vectors')
    plt.xticks(r_values)
    plt.grid()
    plt.show()
    
# Calculate Minkowski distance between two feature vectors in Class 1
calculate_minkowski_distance(df_class1, '0', '1')
```

```python
from sklearn.model_selection import train_test_split
```

```python
def prepare_data(df):
    # Separating features and labels
    x = df.iloc[:, :-1].values  # Feature vectors
    y = df['Judgement Status'].values  # Labels
    return x, y
x,y = prepare_data(df)

def split_data(x, y, test_size=0.3, random_state=42):
    # Splitting the dataset into training and testing sets
    return train_test_split(x, y, test_size=test_size, random_state=random_state)
    
# Split the data
x_train, x_test, y_train, y_test = split_data(x, y)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
def train_knn(x_train, y_train, k=3):
    # Initializing the kNN classifier with specified k
    knn = KNeighborsClassifier(n_neighbors=k)
    # Training the classifier
    knn.fit(x_train, y_train)
    return knn


# Train the kNN model with k=3
knn = train_knn(x_train, y_train, k=3)
```

```python
def evaluate_model(knn, x_test, y_test):
    # Testing the classifier and calculating accuracy
    accuracy = knn.score(x_test, y_test)
    return accuracy

# Evaluate the model
accuracy = evaluate_model(knn, x_test, y_test)
print(f"Accuracy of kNN (k=3): {accuracy * 100:.2f}%")
```

```python
def predict(knn, x_test):
    # Predicting the class labels for the test set
    return knn.predict(x_test)

# Predicting the class labels for the test set
y_pred = predict(knn, x_test)

 # Displaying the first 10 predictions alongside actual labels
for i in range(10):
    print(f"Predicted: {y_pred[i]}, Actual: {y_test[i]}")
```

```python
# Evaluate accuracy for different k values
accuracies = []
k_values = range(1, 12)

for k in k_values:
    knn = train_knn(x_train, y_train, k)
    accuracies.append(evaluate_model(knn, x_test, y_test))

def plot_accuracy(k_values, accuracies):
    # Plotting the accuracy
    plt.figure(figsize=(8, 8))
    plt.plot(k_values, accuracies, marker='o')
    plt.xlabel('k')
    plt.ylabel('Accuracy')
    plt.title('Accuracy of kNN Classifier with varying k')
    plt.grid()
    plt.show()

# Plot accuracy for varying k
plot_accuracy(k_values, accuracies)
```

```python
from sklearn.metrics import confusion_matrix, precision_score, recall_score, f1_score

def calculate_metrics(y_test, y_pred):
    # Confusion Matrix
    conf_matrix = confusion_matrix(y_test, y_pred)
    print(f"Confusion Matrix:\n{conf_matrix}")

    # Precision, Recall, and F1-Score for multiclass
    precision = precision_score(y_test, y_pred, average='macro')
    recall = recall_score(y_test, y_pred, average='macro')
    f1 = f1_score(y_test, y_pred, average='macro')

    print(f"Precision: {precision:.2f}")
    print(f"Recall: {recall:.2f}")
    print(f"F1-Score: {f1:.2f}")

# Calculate and display metrics
calculate_metrics(y_test, y_pred)
```
